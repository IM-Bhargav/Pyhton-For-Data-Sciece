```python
import numpy as np
import pandas as pd
import random 
import math
import matplotlib.pyplot as plt
random.seed(0)
```

```python
def compute_reciprocal(values):
    output = np.empty(len(values))
    for i in range(len(values)):
        output[i] = 1.0/values[i]
    return output
```

```python
values = [random.randint(10,100) for x in range(10)]
compute_reciprocal(values)
```

```python
import matplotlib.pyplot
import numpy as np
from sklearn import datasets,linear_model
from sklearn.metrics import mean_squared_error,r2_score
```

```python
d_X,d_y = datasets.load_diabetes(return_X_y= True)
```

```python
d_X = d_X[:,np.newaxis,2]
```

```python
dXTrain = d_X[:-20]
dXTest = d_X[-20:]
```

```python
dYTrain = d_y[:-20]
dYTest = d_y[-20:]
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(dXTrain,dYTrain)
```

```python
pred = regr.predict(dXTest)
```

```python
pred
```

```python
regr.coef_
```

```python
mean_squared_error(dYTest,dXTest)
```

```python
r2_score(dYTest,dXTest)
```

```python
plt.scatter(dXTest,dYTest,color="black")
plt.plot(dXTest,pred,color="red",linewidth = 0.1)
plt.xticks(())
plt.yticks(())
plt.show()
```
